```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/Arhaan/CABKrittikaTutorials/main/Selection_Assignment/data.txt')
dist=df['mod0'].to_numpy()
vel=df['vgsr'].to_numpy()
B=(dist/5)-5
xdata=np.power(10,B)
print(xdata)
```

```python
%matplotlib inline
# import all the important packages

import numpy as np
from scipy.optimize import curve_fit as cf
import matplotlib.pyplot as plt
```

```python
# Define a new function
def func(x,a,b):
    return a*x + b
ydata = vel 

plt.scatter(xdata,ydata,s=1)
p_opt, p_cov = cf(func,xdata,ydata)
plt.plot(xdata,func(xdata,*p_opt),label='Best Fit',color='black')

plt.scatter(xdata,ydata,label='Original Data',s=1,color='orange')
plt.xlabel('distance of galaxies ')
plt.ylabel('line of sight velocity')
plt.title('Linear Plot')
plt.legend()
plt.scatter(x=xdata,y=df['vgsr'])



```
